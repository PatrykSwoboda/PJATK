```python
from pyspark import SparkConf, SparkContext

conf = SparkConf().setMaster('local').setAppName('zad_1_and_2')
sc = SparkContext(conf=conf)
```

```python
def parse_line(text):
  fields = text.split(',')
  customer_id = int(fields[0])
  val = float(fields[2])
  return customer_id, round(val, 2)

lines = sc.textFile('customer-orders.csv')
rdd_1 = lines.map(parse_line)
```

```python
total_customer = rdd_1.reduceByKey(lambda x, y: round(x+y, 2))

result = total_customer.collect()

for r in result:
  print(r)
```

```python
import collections

sorted_results = collections.OrderedDict(sorted(result))

print('Summarized values by each customer.\n')
for k, v in sorted_results.items():
  print(f'customer {k} -> {v}')
```
